# Gettting the data from ereefs

```python
base_url = "http://thredds.ereefs.aims.gov.au/thredds/dodsC/s3://aims-ereefs-public-prod/derived/ncaggregate/ereefs/GBR4_H2p0_B3p1_Cq3b_Dhnd/daily-monthly/EREEFS_AIMS-CSIRO_GBR4_H2p0_B3p1_Cq3b_Dhnd_bgc_daily-monthly-"

smonth = 12   # Starting month 
year = 2016    # Year

# Based on the server the file naming convention 
biofiles = [f"{base_url}{year}-{month:02}.nc" for month in range(smonth, smonth+1)]


smonth = 1   # Starting month 
emonth = 2   # Starting month 
year = 2017    # Year

# Based on the server the file naming convention 
biofiles2 = [f"{base_url}{year}-{month:02}.nc" for month in range(smonth, emonth+1)]
```

```python
biofiles2
```

```python
biofile = [biofiles[0],biofiles2[0],biofiles2[1]]
biofile
```

```python
lat_lon_list = [(-11.7569,143.4861),(-19.1688,147.3912),(-24.0133,152.7812)]
attr_names = ["CS_bleach","omega_ar"]
month_map = {0:"dec",1:"jan",2:"feb"}
attr_map = {"CS_bleach":"bleach","omega_ar":"omega"}
for i,(lat,lon) in enumerate(lat_lon_list):
    for j,nc_input_name in enumerate(biofile):
        for k,attr_name in enumerate(attr_names):
            ds = xr.open_dataset(biofile[j])

            # Taking the bleach rate
            bleach = ds[attr_name]

            site1_lat = lat
            site1_lon = lon

            # Region around it 5 degrees 
            lon_bnds = [site1_lon-2.5, site1_lon+2.5]
            lat_bnds = [site1_lat-2.5, site1_lat+2.5]

            # Select data in the region
            site1bleach = bleach.sel(latitude=slice(*lat_bnds), longitude=slice(*lon_bnds))

            # Make the average of the variable over the month (bleaching rate)
            meanbleach = site1bleach.mean('time')

            # Find the position of the maximum
            imax = meanbleach.where(meanbleach == meanbleach.max(), drop = True).squeeze()

            # New site position
            nsite1_lat = imax.latitude.values
            nsite1_lon = imax.longitude.values

            print(nsite1_lat,nsite1_lon)

            # Extract the values for the this new site
            site1bleach = site1bleach.sel(latitude=nsite1_lat, longitude=nsite1_lon, method='nearest')
            out_filename = "{}_site{}_{}.nc".format(month_map[j],i+1,attr_map[attr_name])
            print(out_filename)
            site1bleach.to_netcdf(out_filename)
```

```python
all_df = pd.merge(df_2017_8,df2_Oxygen,on=["Aggregated Date/Time","Site Name"])
```

```python
all_df.head()
```

```python
import numpy as np
site1_df=all_df.loc[all_df["Site Name"]=="site 1",]
plt.figure(figsize=(10, 8), dpi=80)
plt.scatter(site1_df["mean"], site1_df["Oxygen"])
z = np.polyfit(site1_df["mean"],site1_df["Oxygen"], 1)
p = np.poly1d(z)
plt.plot(site1_df["mean"],p(site1_df["mean"]))
plt.xlabel("temp(℃)")
plt.ylabel("Oxygen (mol / m-3）")
plt.title("Site1 Temperature and Oxygen")
plt.show()
```
